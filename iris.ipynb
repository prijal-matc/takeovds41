```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
```

```python
data_flower=load_iris()
```

```python
data_flower
```

```python
X=data_flower['data']
y=data_flower['target']
y=to_categorical(y,num_classes=3)
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=.2)
```

```python
print(X.shape)
print(y.shape)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
# Input layer
model.add(Dense(64, input_shape=(4,), activation='relu'))
# hidden layer
model.add(Dense(32, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(28, activation='relu'))
model.add(Dense(26, activation='relu'))
model.add(Dense(24, activation='relu'))
# out layer
# because this is a multiclass classification we use softmax function
model.add(Dense(3, activation='softmax'))
# Model compile
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=15, batch_size=20)
#model.fit(X, y, epochs=20, batch_size=10)
loss_test, accuracy_test = model.evaluate(X_test, y_test)
loss_train, accuracy_train = model.evaluate(X_train, y_train)
print('Accuracy on test set: %.2f' % (accuracy_test*100))
print('Accuracy on train set: %.2f' % (accuracy_train*100))
print("loss test ",loss_test)
print("loss train ",loss_train)
```
